```python
# calculadora simples

# Entrada 2 valores
# processamento calcular usando operadores aritméticos
# Saída Mostrar o resultado



# Entrada

numero1 = float(input("digite um numero:"))
numero2 = float(input("digite o segundo numero:"))


# Processamento

soma = numero1 + numero2
subtração = numero1 - numero2
multiplicação = numero1 * numero2
divisão = numero1 / numero2
div_inteira = numero1 // numero2
resto = numero1 % numero2
potencia = numero1 ** numero2

# Saída

print (f"o resultado da soma de dois numeros é igual: {soma}")
print (f"o resultado da subtração de dois numeros é igual: {subtração}")
print (f"o resultado da multiplicação de dois numeros é igual: {multiplicação}")
print (f"o resultado da divisão de dois numeros é igual: {divisão}")
print (f"o resultado da divisão inteira de dois numeros é igual: {div_inteira}")
print (f"o resultado da soma de dois numeros é igual: {divisão}")
print (f"o resultado do resto de dois numeros é igual: {resto}")
print (f"o resultado da potencia de dois numeros: {potencia}")
```
