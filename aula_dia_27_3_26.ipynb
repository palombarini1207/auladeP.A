```python
# Estrutura de decisões ( if, elf e else)

# Escrever um programa que compara dois numeros

# Entrada
a = int (input("Digite um numero inteiro: "))
b = int (input("Digite um numero inteiro: "))

# Processamento / saída

if a == b:
    print (f"{a} é igual a {b} ")
else:
    print (f"{a} é diferente de {b}")
```

```python
# Escrever um programa que verifica se o numero digitado é par ou ímpar

# Entrada
numero = int ( input("digite um numero: "))

# processamento ? saida

if numero % 2 == 0:

  print (f" o numero {numero} é par")

else:
  print (f" o numero {numero} é impar")
```

```python
# Escrever um programa que o usario digita o nome e a idade e verifica se o usuario é criança , adolecênte ou adulto

# Entrada
nome = str(input("Digite o seu nome: "))
idade = int(input("Digite sua idade: "))

# Processamento / saída

if idade < 12 :
   print (f"{nome} você é uma criança")

elif idade < 18:
  print (f"{nome}, você é um adolescente")

else:
  print (f"{nome}, você é um adulto")
```

```python
# Escrever um programa de uma calculadora simples. O usuario ira digitar 02 numeros e escolhera uma operação aritmetica

# Entrada

print ("----------CALCULADORA----------|")
print ("                               |")
print ("----------MENU-----------------|")
print("                                |")
print("opção = 1 - soma                |")
print("opção = 2 - subtração           |")
print("opção = 3 - multiplicação       |")
print("opção = 4 - divisão             |")
print("--------------------------------|")

# Entrada numero
numero = float (input("digite um numero: "))
opcao = int (input("escolha a opção desejada: "))
numero1 = float (input("digite um numero: "))

# Processamento / saída

if opcao == 1:
  print (f"a soma de {numero} e {numero1} é igual a {numero + numero1}")

elif opcao == 2:
  print (f"a subtração de {numero} e {numero1} é igual a {numero - numero1}") # Adicionado o bloco para a opção 2

elif opcao == 3:
  print (f"a multiplicação de {numero} e {numero1} é igual a {numero * numero1}") # Adicionado o bloco para a opção 3

elif opcao == 4:
  # Adicionado o bloco para a opção 4, com tratamento para divisão por zero
  if numero1 != 0:
    print (f"a divisão de {numero} por {numero1} é igual a {numero / numero1}")
  else:
    print ("Não é possível dividir por zero!")

else:
  print("opção inválida")
```
